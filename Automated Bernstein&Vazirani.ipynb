```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
secret_number = input("Enter the number in bits you want the computer to guess: ")
```

```python
circuit = QuantumCircuit(len(secret_number) + 1 , len(secret_number))
```

```python
circuit.h(range(len(secret_number)))
circuit.x(len(secret_number))
circuit.h(len(secret_number))

circuit.barrier()

for index, one in enumerate(secret_number[::-1]):
    print(f"index {index} is {one}")
    if one == "1":
        circuit.cx(index, len(secret_number))

circuit.barrier()

circuit.h(range(len(secret_number)))

circuit.barrier()

circuit.measure(range(len(secret_number)), range(len(secret_number)))


```

```python
circuit.draw(output = "mpl")
```

```python
from qiskit_aer import AerSimulator
```

```python
simulator = AerSimulator()
job = simulator.run(circuit, shot = 1024)
result = job.result()
```

```python
counts = result.get_counts(circuit)
print(f"Number of counrs is {counts}")
```
